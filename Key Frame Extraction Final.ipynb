```python
def norm_stddev(video_file):
    cap = cv2.VideoCapture(video_file)   
    std_dev = list()
    #print("cap.isOpened()",cap.isOpened())
    f=1
    while cap.isOpened():
        ret, curr_frame = cap.read()
        if ret:
            if f!=1:
                diff = cv2.absdiff(curr_frame, prev_frame)
                #print("diff:",diff)
                diff_norm = np.linalg.norm(diff)
                std_dev.append(diff_norm)
            prev_frame = curr_frame
            f=f+1
        else:
            break
        
    sigma = np.std(std_dev)
    #print("sigma:"+ str(sigma))
    
    cap.release()
    cv2.destroyAllWindows()
    return sigma
```

```python
def extract_key_images(file_name, video_file, output_dir, std_dev, n):
    
    cap = cv2.VideoCapture(video_file)
    key_images = list()
    #print("cap.isOpened()",cap.isOpened())
    f=1
    while cap.isOpened():
        ret, curr_frame = cap.read()
        #print("ret:", ret)
        if ret:
            if f!=1:
                #print("Inside 2:", f)
                diff = cv2.absdiff(curr_frame, prev_frame)
                #print("diff:",diff)
                diff_norm = np.linalg.norm(diff)
                #print("diff_norm:",diff_norm)
                if diff_norm > (n * std_dev):              
                    imageName = file_name+'_'+"image{}.jpg".format(f)
                    cv2.imwrite(output_dir+'/'+imageName,curr_frame)
                    key_images.append(imageName)
            if f == 1:
                #print("Inside 1:", f)
                imageName = file_name+'_'+"image{}.jpg".format(f)
                cv2.imwrite(output_dir+'/'+imageName,curr_frame)
                key_images.append(imageName)
            prev_frame = curr_frame
            f =f+1
        else:
            break
    
    cap.release()
    cv2.destroyAllWindows()
    #print("key_images inside extract_key_images:",key_images)
    return key_images
```

```python
def extract_key_frames(file, videoFile, outputFolderName):
    print("file:",file)
    count = 0
    key_images = list()
    cap = cv2.VideoCapture(videoFile)   # capturing the video from the given path
    frameRate = cap.get(5) #frame rate
    x=1
    while(cap.isOpened()):
        frameId = cap.get(1) #current frame number
        #print("frameId:",frameId)
        ret, frame = cap.read()
        if (ret != True):
            break
        if (frameId % math.floor(frameRate) == 0):
            filename =file+'_'+"frame%d.jpg" % count;count+=1
            print("filename:",filename)
            cv2.imwrite(outputFolderName+'/'+filename, frame)
            key_images.append(filename)
    cap.release()
    print ("Done!")
    return key_images
```
